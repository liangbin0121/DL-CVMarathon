思考一下我們前面有提到圖片是矩陣，但維度可能會不一樣
例如灰階圖只有兩個維度，RGB 彩圖則有 3 個維度

假如今天我們把 RGB 3 個維度拆開來看會有甚麼不同的效果呢？

```python
import matplotlib.pyplot as plt # plt 用於顯示圖片
import matplotlib.image as mpimg # mpimg 用於讀取圖片
import numpy as np

%matplotlib inline
img_path = 'lena.png'

lena = mpimg.imread(img_path) # 讀取和程式碼處於同一目錄下的 lena.png

# 此時 lena 就已經是一個 np.array 了，可以對它進行任意處理
lena.shape #(512, 512, 3) #以彩色圖片的方式載入
plt.imshow(lena) # 顯示圖片
plt.axis('off') # 不顯示座標軸
plt.show()
```

```python
def rgb2gray(rgb):
 return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
gray = rgb2gray(lena)  

plt.imshow(gray, cmap='Greys_r')
plt.axis('off')
plt.show()
```

```python
# R
print(lena[:,:,0],lena[:,:,0].shape)
# G
print(lena[:,:,1],lena[:,:,1].shape)
# B
print(lena[:,:,2],lena[:,:,2].shape)

 
fig,axes = plt.subplots(1,3,figsize=(12,7))
axes[0].imshow(lena[:,:,0])
axes[0].set_title('R')
axes[1].imshow(lena[:,:,1])
axes[1].set_title('G')
axes[2].imshow(lena[:,:,2])
axes[2].set_title('B')
 
plt.show()
```
